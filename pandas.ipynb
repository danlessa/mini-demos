```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.set()
```

```python
N = 100

dict_data = {'x': np.random.rand(N),
             'y_1': np.random.randn(N),
             'y_2': np.random.randn(N),
             'y_3': np.random.randn(N)
            }

data = (pd.DataFrame(dict_data)
          .assign(y_ratio=lambda df: df.y_2 / df.y_1)
          .assign(above_quantile=lambda df: df.y_3 > df.y_3.quantile(0.7))
       )
```

```python
data_upper = data.query('x > 0.5 & above_quantile == True')
sns.lineplot(data=data_upper,
             x='x',
             y='y_ratio')
```

```python
sns.distplot(data.y_ratio)
```
